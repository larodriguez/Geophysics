## Grupo A

El trabajo del grupo A, fue se tres lineas diferentes. Dichas lineas se realizaron en un espacio relativamente corto. La geologia de la zona es sedimentaria principalmente depositos cuaternarios, dichas rocas vienen de rocas igneas y metamorficas de la cordillera central del pais. 

Linea 1: Coordenadas: 4°59'31''N 74°54'56''W

Linea 2: Coordenadas: 4°58'39''N 74°54'42''W

Linea 3: Coordenadas: 4°58'39''N 74°54'42''W

```python
plt.figure(figsize=(20, 16))
plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.subplot(231)
plt.plot(x3, y3, c='r')
plt.title(r'$Linea 1, Grupo A$',fontsize=20)
plt.xlabel(r'$Eje X$')
plt.ylabel(r'$Eje y$')

plt.subplot(232)
plt.plot(x4, y4, c='g')
plt.title(r'$Linea 2, Grupo A$',fontsize=20)
plt.xlabel(r'$Eje X$')
plt.ylabel(r'$Eje y$')

plt.subplot(233)
plt.plot(x5, y5, c='g')
plt.title(r'$Linea 3, Grupo A$',fontsize=20)
plt.xlabel(r'$Eje X$')
plt.ylabel(r'$Eje y$')
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1, 3, 1, projection='3d')

p = ax.plot_trisurf(x3, y3, z3, cmap=cm.jet, linewidth=0.2)
cb = fig.colorbar(p)
ax.set_title(r'$Linea 1, Grupo A$',fontsize=20)
ax.set_xlabel(r'$Eje X$', fontsize=15)
ax.set_ylabel(r'$Eje Y$', fontsize=15)
ax.set_zlabel(r'$Anomalia Magnetica (nT)$', fontsize=15)
plt.grid() 

ax = fig.add_subplot(1, 3, 2, projection='3d')

p1 = ax.plot_trisurf(x4, y4, z4, cmap=cm.jet, linewidth=0.2)
cb = fig.colorbar(p1)
ax.set_title(r'$Linea 2, Grupo A$',fontsize=20)
ax.set_xlabel(r'$Eje X$', fontsize=15)
ax.set_ylabel(r'$Eje Y$', fontsize=15)
ax.set_zlabel(r'$Anomalia Magnetica (nT)$', fontsize=15)
plt.grid() 

ax = fig.add_subplot(1, 3, 3, projection='3d')

p1 = ax.plot_trisurf(x5, y5, z5, cmap=cm.jet, linewidth=0.2)
cb = fig.colorbar(p1)
ax.set_title(r'$Linea 3, Grupo A$',fontsize=20)
ax.set_xlabel(r'$Eje X$', fontsize=15)
ax.set_ylabel(r'$Eje Y$', fontsize=15)
ax.set_zlabel(r'$Anomalia Magnetica (nT)$', fontsize=15)
plt.grid() 

plt.show()
```

La forma de las cuadricualas realizadas no fue muy uniforme, de igual manera los datos obetenidos se muestran muy planos donde los contrastes de magnetismo no se pueden diferenciar con facilidad. Se puede ver en las tres graficas que los valores de las suceptibilidades magneticas es positivo, lo cual es correlacionable con la litologia que tiene la zona de estudio.
